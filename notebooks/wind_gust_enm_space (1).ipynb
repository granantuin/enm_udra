```python
ria_guss=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Marin/ria_enm_windplus.csv").drop(columns=["time","Unnamed: 0","dir_udr","spdd_udr","dird_udr","spd_udr"])
ria_guss
```
